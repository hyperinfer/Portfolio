## 1.1 BoW

```python
X_train, X_test, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=2805)
```

```python
#We conduct a hyperparameter gridsearch to ensure that we are using the best possible specification

logit = LogisticRegression(max_iter = 1000)

logit_parameters = {
    "solver" : ("lbfgs", "liblinear", "newton-cg", "newton-cholesky", "sag", "saga"),
    "penalty" : ("l1", "l2", "elasticnet", None)
}

logit_clf = GridSearchCV(logit, logit_parameters, n_jobs = -1)

logit_clf.fit(X_train, y_train)


with open("logit_bow_clf.txt", "w") as f:
    f.write(str(logit_clf.best_params_))
    f.close()
```
